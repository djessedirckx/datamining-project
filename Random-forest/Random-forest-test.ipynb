```python
import pandas as pd
training_data = "../Data/training_binary.csv"
data = pd.read_csv(training_data)
y = data['cuisine']
```

```python
x_frame = data.drop(['cuisine', 'id'],axis=1)
x_frame.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from Toolbox.treeprint import tree_print

dtc = DecisionTreeClassifier(min_samples_split=10)
dtc = dtc.fit(x_frame,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_frame,y)
sum(clf.predict(x_frame) == y)
```

```python
predictions = clf.predict(x_frame)
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y,predictions,labels=list(set(y))),index=list(set(y)),columns=list(set(y)))
```

```python
import pandas as pd
test_data = "../Data/test_binary.csv"
data = pd.read_csv(test_data)
```

```python
from sklearn.ensemble import RandomForestClassifier
y = data['cuisine']
x_frame = data.drop(['cuisine','id'],axis=1)

waardes = []
for i in range(2,400,20):
    print(i)
    clf = RandomForestClassifier(criterion = 'gini', max_features = 'auto',n_estimators = 60, max_depth=i)
    clf.fit(x_frame,y)
    waardes.append(sum(clf.predict(x_frame)==y))
```

```python
plt.plot(range(2,400,20),waardes)
plt.grid()
```

```python
import matplotlib.pyplot as plt
predictions = clf.predict(x_frame)
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y,predictions,labels=list(set(y))),index=list(set(y)),columns=list(set(y)))
```
