```python
import pandas as pd
training_data = "../Data/training_binary.csv"
data = pd.read_csv(training_data)
y = data['cuisine']
```

```python
x_frame = data.drop(['cuisine', 'id'],axis=1)
x_frame.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from Toolbox.treeprint import tree_print

dtc = DecisionTreeClassifier(min_samples_split=10)
dtc = dtc.fit(x_frame,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_frame,y)
sum(clf.predict(x_frame) == y)
```

```python
predictions = clf.predict(x_frame)
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y,predictions,labels=list(set(y))),index=list(set(y)),columns=list(set(y)))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [200, 500],
    'max_features': ['auto','sqrt','log2'],
    'max_depth': [4,5,6,7,8],
    'criterion': ['gini','entropy']
}
gridsearch = GridSearchCV(estimator = clf, param_grid=param_grid, cv=5)
gridsearch.fit(x_frame,y)
```
