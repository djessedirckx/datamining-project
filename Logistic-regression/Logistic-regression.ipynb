```python
# Load training data
import pandas as pd
training_data = "../Data/training_binary.csv"
data = pd.read_csv(training_data)
y_train = data['cuisine']

x_train = data.drop(['cuisine', 'id'],axis=1)
x_train.head()
```

```python
# Load test data
test_data = "../Data/test_binary.csv"
data = pd.read_csv(training_data)
y_test = data['cuisine']

x_test = data.drop(['cuisine', 'id'],axis=1)
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train, y_train)
```

```python
training_predictions = clf.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix

cm = pd.DataFrame(confusion_matrix(y_train,training_predictions,labels=list(set(y_train))),index=list(set(y_train)),columns=list(set(y_train)))
cm
```

```python
# Performance results using default parameters

from sklearn.metrics import precision_score, recall_score, accuracy_score

print("Accuracy: {}".format(clf.score(x_train, y_train)))
print("Number of inequalities: {}\n".format((y_train!=training_predictions).sum()))

# TP rate
print("Average TP rate: {}".format(precision_score(y_train, training_predictions, average='micro')))
```
